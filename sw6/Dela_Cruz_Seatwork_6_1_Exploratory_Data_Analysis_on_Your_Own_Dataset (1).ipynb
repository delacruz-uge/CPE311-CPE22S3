```python
filepath = '/content/LeaguePlayerStats.csv'
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(filepath)
```

```python
# total number of players
num_record = len(data)
print("Total number of players: ", num_record)
```

```python
from matplotlib import pyplot as plt
data['winRate'].plot(kind='hist', bins=40, title='winRate')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#  Display the first 20 records
data[0:20]
```

```python
# Display the last 20 records
data[-20:]
```

```python
# Changing column name
data.rename(columns={'summonerName': 'In Game Name'}, inplace = True)
data
```

```python
# Create a new dataframe with a player with no turret take downs
noturretdf = pd.DataFrame(data)
noturretdf = data[data['turretTakedowns'] == 0]
noturretdf
```

```python
# graph of player's average losses with no turret takedowns
from matplotlib import pyplot as plt
noturretdf['losses'].plot(kind='hist', bins=40, title='losses')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Create a new dataframe that is level 900 above
leveldf = pd.DataFrame(data)
leveldf = data[data['summonerLevel'] >= 900].copy()
leveldf
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
leveldf.groupby('rank').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#line graph of each player's win
from matplotlib import pyplot as plt
leveldf['wins'].plot(kind='line', figsize=(10, 6), title='Wins')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# Making another dataframe showing "kills" and the Account info only
killdf = pd.DataFrame(data, columns=['In Game Name', 'kills', 'rank', 'summonerLevel'])
killdf
```

```python
# graph for average kills with each player's level
from matplotlib import pyplot as plt
killdf.plot(kind='scatter', x='kills', y='summonerLevel', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# graph of each player's average wins
from matplotlib import pyplot as plt
data['totalDamageDealt'].plot(kind='line', figsize=(10, 6), title='Average Damage dealt')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# getting the mean of kills and deaths of each player
mean_killndeath = np.mean(data['kills']), np.mean(data['deaths'])
mean_killndeath
```

```python
# getting the median of wins and summonerLevel of each player
median_killndeath = np.median(data['wins']), np.median(data['summonerLevel'])
median_killndeath
```

```python
# Getting the Min and Max goldEarned of each player
max_goldearned = max(data['goldEarned'])
min_goldearned = min(data['goldEarned'])

print("Max gold earned:",max_goldearned)
print("Min gold earned:",min_goldearned)
```

```python
# Use numpy to get the standard deviation of the totalDamageDealt.
stdev_totalDamageDealt = np.std(data['totalDamageDealt'])
stdev_totalDamageDealt
```
